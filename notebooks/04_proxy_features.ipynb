```python
import pandas as pd
import numpy as np

from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize

import textstat
```

```python
df = pd.read_csv("../outputs/clean_news.csv")

df.head()
```

```python
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
```

```python
df["clean_text"] = df["clean_text"].astype(str)
```

```python
def get_emotion_score(text):
    
    text = str(text)   # convert safely
    
    score = sia.polarity_scores(text)
    
    return abs(score['compound'])
```

```python
df["emotion_score"] = df["clean_text"].apply(get_emotion_score)
```

```python
df = df.dropna(subset=["clean_text"])
```

```python
df = df.iloc[:500]
```

```python
urgent_words = [
    "breaking", "urgent", "alert",
    "shocking", "immediately",
    "update", "crisis"
]

def get_urgency_score(text):
    text = str(text)
    return sum(word in text for word in urgent_words)
```

```python
from nltk.tokenize import word_tokenize

def lexical_diversity(text):
    
    words = word_tokenize(str(text))
    
    return len(set(words)) / len(words) if len(words) > 0 else 0
```

```python
df["emotion_score"] = df["clean_text"].apply(get_emotion_score)

df["urgency_score"] = df["clean_text"].apply(get_urgency_score)

df["length_score"] = df["clean_text"].apply(lambda x: len(x.split()))

df["lexical_diversity"] = df["clean_text"].apply(lexical_diversity)

df["readability"] = df["clean_text"].apply(textstat.flesch_reading_ease)
```

```python
df.to_csv("../outputs/proxy_features.csv", index=False)
```
