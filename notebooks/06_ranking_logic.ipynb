```python
import pandas as pd
```

```python
df = pd.read_csv("../outputs/final_scored_data.csv")

df.head()
```

```python
ranked_df = df.sort_values(
    by="popularity_score",
    ascending=False
)
```

```python
ranked_df["rank"] = range(1, len(ranked_df) + 1)
```

```python
ranked_df[["title", "popularity_score", "rank"]].head(10)
```

```python
ranked_df.to_csv("../outputs/ranked_news.csv", index=False)
```

```python
top5 = ranked_df.head(5)

for i, row in top5.iterrows():
    print(f"Rank {row['rank']}:")
    print("Title:", row["title"])
    print("Score:", row["popularity_score"])
    print("-" * 50)
```
