```python
import pandas as pd
import numpy as np

from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize

import textstat
```

```python
df = pd.read_csv("../outputs/clean_news.csv")

df.head()
```

```python
df = pd.read_csv("../outputs/proxy_features.csv")

import numpy as np
embeddings = np.load("../outputs/embeddings.npy")

df.shape, embeddings.shape
```

```python
emb_df = pd.DataFrame(embeddings)

final_df = pd.concat([df.reset_index(drop=True), emb_df], axis=1)
```

```python
final_df["popularity_score"] = (
    0.3 * final_df["emotion_score"] +
    0.2 * final_df["urgency_score"] +
    0.2 * final_df["lexical_diversity"] +
    0.15 * final_df["readability"] +
    0.15 * final_df["length_score"]
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

final_df["popularity_score"] = scaler.fit_transform(
    final_df[["popularity_score"]]
)
```

```python
final_df.to_csv("../outputs/final_scored_data.csv", index=False)
```
