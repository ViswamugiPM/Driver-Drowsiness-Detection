🎯 Mentor Question

Answer (say like this):

The project constraint says we must work only with raw text.
So we keep only title and description because they contain the semantic content needed for Transformer representation learning.

```python
df = df[["title", "description"]]

df.head()
```

🎯 Mentor Question
❓ why combine title and description?

Answer:

Title contains attention-oriented words and description contains context.
Combining both gives full semantic input for the Transformer model.

```python
df["text"] = df["title"] + " " + df["description"]

df.head()
```

```python
import re

def clean_text(text):

    text = str(text)

    text = text.lower()

    text = re.sub(r"http\S+", "", text)

    text = re.sub(r"[^a-zA-Z ]", "", text)

    text = re.sub(r"\s+", " ", text)

    return text


df["clean_text"] = df["text"].apply(clean_text)

df.head()
```

```python
df.to_csv("../outputs/clean_news.csv", index=False)
```
