```python
class_map = {'Closed': 0, 'Open': 1, 'no_yawn': 2, 'yawn': 3}
```

```python
def get_score(pred_label):
    if pred_label == "Closed":
        return 2
    elif pred_label == "yawn":
        return 1
    else:
        return 0
```

```python
import random

sample_predictions = ["Open", "Open", "yawn", "yawn", 
                      "Closed", "Closed", "Closed", 
                      "yawn", "Open", "Open"]
```

```python
scores = [get_score(p) for p in sample_predictions]

print(scores)
```

```python
def get_stage(score):
    if score == 0:
        return "Alert"
    elif score == 1:
        return "Mild Fatigue"
    else:
        return "Severe Fatigue"

stages = [get_stage(s) for s in scores]
print(stages)
```

```python
import os

os.makedirs("outputs", exist_ok=True)
```

```python
import matplotlib.pyplot as plt

plt.plot(scores, marker='o')
plt.title("Driver Fatigue Progression")
plt.xlabel("Time (Frames)")
plt.ylabel("Fatigue Level (0=Alert, 1=Mild, 2=Severe)")
plt.grid()

# Save output
plt.savefig("outputs/fatigue_curve.png")

plt.show()
```
